```python
#seasons for the data --> slower in summer that winter?
#spring 3-5 , summer 6-8, fall 9-11, winter 12-2
#split between 2 decimals
```

```python
df3['race_month'] = df3['race_day'].str.split('.').str.get(1).astype(int)
```

```python
df3.head(25)
```

```python
df3['race_season'] = df3['race_month'].apply(lambda x: 'winter' if x > 11 
                                             else 'fall' if x > 8 
                                             else 'summer' if x > 5 
                                             else 'spring' if x > 2 
                                             else 'winter')
```

```python
df3.head(25)
```

```python
df3.groupby('race_season')['race_number_of_finish'].agg(['mean','count']).sort_values('mean', ascending = False)
```

```python
#50 miler only
df3.query('race_distance == "50mi"').groupby('race_season')['race_number_of_finish'].agg(['mean','count']).sort_values('mean', ascending = False)
```
